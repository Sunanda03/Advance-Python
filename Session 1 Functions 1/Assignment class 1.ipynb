```python
def factorial(n):
    if n==0:
        f = 1
    else:
        f = 1
        for i in range(1,n+1):
            f = f*i
        return f
```

### Q4 : For given P INR, N in Years and R in %p.a. calculate EMI (Equal Monthly Installements)
![Alt text](image.png)

1. 1 year = 12 %
2. 12 month = 12 %
3. 1 month = (12)%/12 = 1% = 1/100 = 0.01 per month

```python
def calculate_EMI(P, N, R):
    # convert from years to months
    n = N*12

    # Convert R from %p.a. to %p.m.
    r = R/(12*100)

    # EMI Calculater
    x = (1+r)**n
    EMI = P*r*x/(x-1)

    return round(EMI)
```

```python
e1 = calculate_EMI(P=500000, N=3, R=12.5)
```

```python
e1
```

```python
e2 = calculate_EMI(P=800000, N=5, R=15)
```

### Q5. Take marks of multiple students as input and provide grade accordingly
1. More than 90 : A
2. 80 - 89 : B
3. 70 - 79 : C
4. 60 - 69 : D
5. Below 60 : F

```python
def get_grades(*marks):
    gr = []
    for i in marks:
        if i>=90:
            gr.append('A')
        elif i>=80 and i<=89:
            gr.append('B')
        elif i>=70 and i<=79:
            gr.append('C')
        elif i>=60 and i<=69:
            gr.append('D')
        else:
            gr.append('F')
    return gr
        
```

```python
get_grades(78)
```

```python
g1 = get_grades(65, 98, 75, 82, 86, 30, 55)
g1
```

```python
def analyze_marks(*marks):
    # Find average of marks
    s = sum(marks)
    c = len(marks)
    avg = s/c

    # Find highest marks
    high = max(marks)

    # Find Lowest marks
    low = min(marks)
    
    # Number of students passed and failed
    p = 0
    f = 0
    for i in marks:
        if i>=60:
            p = p + 1
        else:
            f = f + 1

    # Create a dictionary
    dct = {'Average':[avg],
           'Highest':[high],
           'Lowest':[low],
           'nPass':[p],
           'nFail':[f]}
    
    # Covert dct to dataframe
    import pandas as pd
    df = pd.DataFrame(dct)

    return df
```

```python
analyze_marks(35, 78, 93, 85, 86, 62, 56, 65)
```
